```python
import pandas as pd
import requests
import json
from pprint import pprint
import numpy as np
from pandas.io.json import json_normalize
```

```python
#link to The Movies Data Base API 
url = "https://api.themoviedb.org/3/movie/upcoming?api_key=8bc5a5245af0e94fde2d2ac84dffb856&language=en-US&page=1"

```

```python
#creating a variable to store the results found in the url above
response = requests.get(url) 
#storing the responses to a new variable, data
data = response.json()
# pprint(data) 
```

```python
#create variables to store the information from the json into a dataframe
id = []
title = []
popularity = []
vote_average = []
vote_count = []
original_language = []
release_date = []
```

```python
#create for loop that reads through each page of the json response 
num_page = 1
for num_page in range(1, data['total_pages']):
    url = "https://api.themoviedb.org/3/movie/upcoming?api_key=8bc5a5245af0e94fde2d2ac84dffb856&language=en-US&page=1"
    response = requests.get(url).json()

    #for results, append select data to the empty variable arrays above 
    try:
        for x in range(0, len(response['results'])):
            id.append(response['results'][x]['id'])
            popularity.append(response['results'][x]['popularity'])
            vote_average.append(response['results'][x]['vote_average'])
            vote_count.append(response['results'][x]['vote_count'])
            original_language.append(response['results'][x]['original_language'])
            release_date.append(response['results'][x]['release_date'])
   #skip over values that don't exist
    except (KeyError, IndexError):
        print("value doesn't exist")
        
#normalize the json to make sure all variable arrays are the same length and thus can be stored in a pandas dataframe together        
df = pd.json_normalize(response['results'])
df.head(3)
```

```python
#reorganized the columns of the new dataframe 
new_releases_df = df[['id','title','popularity','vote_count','vote_average','release_date','original_language','genre_ids','overview']]
```
