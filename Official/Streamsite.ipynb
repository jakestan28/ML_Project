```python
import requests
import pandas as pd
import random
from pprint import pprint
from config import api_key
```

```python
movie_data = pd.read_csv("now_playing.csv")
```

```python
ids = []

id_data = pd.read_csv("now_playing.csv")
ids = id_data.np_id.tolist()

rand_ids = ids
len(rand_ids)
```

```python
url = "https://utelly-tv-shows-and-movies-availability-v1.p.rapidapi.com/idlookup"

headers = {
    # 'x-rapidapi-key': api_key,
    'x-rapidapi-host': "utelly-tv-shows-and-movies-availability-v1.p.rapidapi.com"
    }

sites = []

for x in rand_ids:
    querystring = {"source_id": x,"source":"tmdb","country":"us"}
    response = requests.request("GET", url, headers=headers, params=querystring).json()

    try:
        response['collection']['name']
        for x in range(0, len(response['collection']['locations'])):
            sites.append(f"{response['id']},{response['collection']['name']},{response['collection']['locations'][x]['name']}")
            
    except(KeyError):
        print(x)
```

```python
tmdb_id = []
movie = []
location = []

for x in sites:
    y = x.split(',')
    tmdb_id.append(y[0])
    movie.append(y[1])
    location.append(y[2])
```

```python
sites_df = pd.DataFrame({"TMDB_ID": tmdb_id, "Movies": movie, "Sites": location}, index=tmdb_id)
```

```python
streaming_df = pd.DataFrame({"TMDB_ID": tmdb_id,"Movies": movie}, index=tmdb_id)
streaming_df = streaming_df.drop_duplicates()
```

```python
for index, row in sites_df.iterrows():
    if row["Sites"] == 'GooglePlayIVAUS':
        streaming_df.at[index, "Google_Play"] = 1
    elif row["Sites"] == 'iTunesIVAUS':
        streaming_df.at[index, "Itunes"] = 1
    elif row["Sites"] == 'AmazonInstantVideoIVAUS':
        streaming_df.at[index, "Amazon_Instant"] = 1
    elif row["Sites"] == 'NetflixIVAUS':
        streaming_df.at[index, "Netflix"] = 1
    elif row["Sites"] == 'HuluIVAUS':
        streaming_df.at[index, "Hulu"] = 1
    elif row["Sites"] == 'HBOMaxIVAUS':
        streaming_df.at[index, "HBO_Max"] = 1
    elif row["Sites"] == 'AmazonPrimeVideoIVAUS':
        streaming_df.at[index, "Amazon_Prime"] = 1
    elif row["Sites"] == 'DisneyPlusIVAUS':
        streaming_df.at[index, "Disney_Plus"] = 1
    else:
        streaming_df.at[index, "Others"] = 1 
        print(row["Sites"])
```

```python
streaming_df = streaming_df.fillna(0)
streaming_df
```
